```python
import matplotlib.pyplot as plt
import numpy as np
def f1(x):
    return x**2
def f2(x):
    return x**0.5
x = np.array(range(0,11))
y = f1(x)
y1 = f2(x) 

plt.plot(x,y,'r-o')
plt.plot(x,y1,"b-.*")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(["X = $x^2$","X = $\sqrt{x}$"])
plt.title("Plot of y vs X")
plt.show()
```

# 2


```python
from math import sin,cos,pi
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
    return sin(x)
def f2(x):
    return cos(x)

x = np.array(np.linspace(0,2*pi,20))

plt.figure(figsize=(12,12))
plt.subplot(2,2,1)
plt.plot(x,[sin(x) for x in x],"r-")
plt.title("y = sin(x)")


plt.subplot(2,2,2)
plt.plot(x,[cos(x) for x in x],"g-")
plt.title("y = cos(x)")


plt.subplot(2,2,3)
plt.scatter(np.random.uniform(15,20,size = 100),np.random.uniform(10,12,size = 100))
plt.title("Scatter-Plot")


plt.subplot(2,2,4)
frequency = np.random.randint(20,25,size = 4)
model = ['M1','M2','M3','M4']
plt.bar(model,frequency,width=0.67)

plt.suptitle('Four Different subplots in one unified plot')
plt.show()
```

Seaborn and ggplot are high level plotting library in python unlike matplotlib ehich is a low level plotting library.

```python
# %pip install pandas
# %pip install seaborn
import pandas as pd
import seaborn as sns
df = pd.DataFrame({'Lalitude':np.random.normal(28,2,size = 200),'Longitude':np.random.normal(85,2,size=200)})
sns.scatterplot(df, x ='Lalitude',y = 'Longitude')
plt.title("Habibi's Access Point")
```

```python
# Using ggplot
#%pip install plotnine
from plotnine import ggplot,aes,geom_point,ggtitle
ggplot(df,aes('Lalitude','Longitude')) + geom_point() + ggtitle("Habibi's Access Point")
```
