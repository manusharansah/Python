```python
import matplotlib.pyplot as plt
import numpy as np
def f1(x):
    return x**2
def f2(x):
    return x**0.5
x = np.array(range(0,11))
y = f1(x)
y1 = f2(x) 

plt.plot(x,y,'r-o')
plt.plot(x,y1,"b-.*")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(["X = $x^2$","X = $\sqrt{x}$"])
plt.title("Plot of y vs X")
plt.show()
```

```python
from math import sin,cos,pi
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
    return sin(x)
def f2(x):
    return cos(x)

x = np.array(np.linspace(0,2*pi,20))

plt.figure(figsize=(12,12))
plt.subplot(2,2,1)
plt.plot(x,[sin(x) for x in x],"r-")
plt.title("y = sin(x)")


plt.subplot(2,2,2)
plt.plot(x,[cos(x) for x in x],"g-")
plt.title("y = cos(x)")


plt.subplot(2,2,3)
plt.scatter(np.random.uniform(15,20,size = 100),np.random.uniform(10,12,size = 100))
plt.title("Scatter-Plot")


plt.subplot(2,2,4)
frequency = np.random.randint(20,25,size = 4)
model = ['M1','M2','M3','M4']
plt.bar(model,frequency,width=0.67)

plt.suptitle('Four Different subplots in one unified plot')
plt.show()
```

# 3
Seaborn and ggplot are high level plotting library in python unlike matplotlib ehich is a low level plotting library.

```python
# %pip install pandas
# %pip install seaborn
import pandas as pd
import seaborn as sns
df = pd.DataFrame({'Lalitude':np.random.normal(28,2,size = 200),'Longitude':np.random.normal(85,2,size=200)})
sns.scatterplot(df, x ='Lalitude',y = 'Longitude')
plt.grid()
plt.title("Habibi's Access Point")
```

```python
# Using ggplot
#%pip install plotnine
from plotnine import ggplot,aes,geom_point,ggtitle
ggplot(df,aes('Lalitude','Longitude')) + geom_point() + ggtitle("Habibi's Access Point")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
data = np.random.normal(200,3,size = 1000)

kde = gaussian_kde(data)
x = np.linspace(min(data),max(data),1000)

plt.hist(data,30,density=True)
plt.plot(x,kde(x),color = 'red', label = 'Normal Fit')

plt.title('Normal Distribution')
plt.legend()
plt.show()
```

```python
# using norm.pdf
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = np.random.normal(200,2,size = 1000)
x = np.linspace(min(data),max(data),1000)
m,sd = np.mean(data),np.std(data)
p = norm.pdf(x,m,sd)

plt.hist(data,30, density = True)
plt.plot(x,p,color = 'red',label = "Normal Fit")
plt.legend()
plt.title("Normal Distribution")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

scores = np.random.randint(80,95,5)
labels = ['M1','M2','M3','M4','M5']
bars = plt.bar(labels,scores)
plt.title("Accuracy of Models")
plt.xlabel("Model")
plt.ylabel("Accuracy")
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x()+ bar.get_width()/2,height,str(height),ha = 'center', va = 'bottom')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = range(1,13)
data1 = np.random.randint(3000,4500,12)
data2 = np.random.randint(5000,7000,12)
data3 = np.random.randint(8500,10500,12)

plt.fill_between(x,data1,color= 'red', alpha = 0.5)
plt.fill_between(x,data2,color= 'green', alpha = 0.5)
plt.fill_between(x,data3,color= 'orange', alpha = 0.5)

plt.plot(x,data1)
plt.plot(x,data2)
plt.plot(x,data3)

plt.legend(['A','B','C'],bbox_to_anchor = (1,1))
plt.title("Sales of different products in different month")
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()

```
