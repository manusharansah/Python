```python
import matplotlib.pyplot as plt
import numpy as np
def f1(x):
    return x**2
def f2(x):
    return x**0.5
x = np.array(range(0,11))
y = f1(x)
y1 = f2(x) 

plt.plot(x,y,'r-o')
plt.plot(x,y1,"b-.*")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(["X = $x^2$","X = $\sqrt{x}$"])
plt.title("Plot of y vs X")
plt.show()
```

```python
from math import sin,cos,pi
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
    return sin(x)
def f2(x):
    return cos(x)

x = np.array(np.linspace(0,2*pi,20))

plt.figure(figsize=(12,12))
plt.subplot(2,2,1)
plt.plot(x,[sin(x) for x in x],"r-")
plt.title("y = sin(x)")


plt.subplot(2,2,2)
plt.plot(x,[cos(x) for x in x],"g-")
plt.title("y = cos(x)")


plt.subplot(2,2,3)
plt.scatter(np.random.uniform(15,20,size = 100),np.random.uniform(10,12,size = 100))
plt.title("Scatter-Plot")


plt.subplot(2,2,4)
frequency = np.random.randint(20,25,size = 4)
model = ['M1','M2','M3','M4']
plt.bar(model,frequency,width=0.67)

plt.suptitle('Four Different subplots in one unified plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
data = np.random.normal(200,3,size = 1000)

kde = gaussian_kde(data)
x = np.linspace(min(data),max(data),1000)

plt.hist(data,30,density=True)
plt.plot(x,kde(x),color = 'red', label = 'Normal Fit')

plt.title('Normal Distribution')
plt.legend()
plt.show()
```

```python
# using norm.pdf
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = np.random.normal(200,2,size = 1000)
x = np.linspace(min(data),max(data),1000)
m,sd = np.mean(data),np.std(data)
p = norm.pdf(x,m,sd)

plt.hist(data,30, density = True)
plt.plot(x,p,color = 'red',label = "Normal Fit")
plt.legend()
plt.title("Normal Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = range(1,13)
data1 = np.random.randint(3000,4500,12)
data2 = np.random.randint(5000,7000,12)
data3 = np.random.randint(8500,10500,12)

plt.fill_between(x,data1,color= 'red', alpha = 0.5)
plt.fill_between(x,data2,color= 'green', alpha = 0.5)
plt.fill_between(x,data3,color= 'orange', alpha = 0.5)

plt.plot(x,data1)
plt.plot(x,data2)
plt.plot(x,data3)

plt.legend(['A','B','C'],bbox_to_anchor = (1,1))
plt.title("Sales of different products in different month")
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()

```

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd

iris = datasets.load_iris()
df = pd.DataFrame(data = iris.data,columns=iris.feature_names)
df['species'] = iris.target
species_name = iris.target_names
for i,speci in enumerate(species_name):
    curr_spe = df[df['species']==i]
    plt.scatter(curr_spe['sepal width (cm)'],curr_spe['sepal length (cm)'],label = speci,cmap='nipy_spectral')
plt.colorbar()
plt.legend()
plt.xlabel('Sepal width(cm)')
plt.ylabel('Sepal length(cm)')
plt.title('Sepal length(cm) vs Sepal width(cm) in Iris dataset')
plt.show()
```

```python
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from matplotlib.patches import Patch
import matplotlib.pyplot as plt


X,Y = make_moons(n_samples=1000,noise=0.2)
x_train,x_test,y_train,y_test = train_test_split(X,Y)
model = SVC(kernel='rbf')
model.fit(x_train,y_train)

font = {'family':'Times New Roman','fontsize':20}
plt.figure(figsize=(12,12))
x_min,x_max = X[:,0].min()-0.5,X[:,0].max()+0.5
y_min,y_max = X[:,1].min()-0.5,X[:,1].max() +0.5

xx,yy = np.meshgrid(np.linspace(x_min,x_max,500),np.linspace(y_min,y_max,500))
z = model.predict(np.c_[xx.ravel(),yy.ravel()])
z = z.reshape(xx.shape)

plt.contourf(xx,yy,z,cmap = 'bwr')
plt.scatter(X[:,0],X[:,1],c = Y,cmap='viridis',edgecolors= 'black')
plt.xlabel("First Feature",fontdict=font)
plt.ylabel("Second Feature",fontdict=font)
plt.title("Decision Boundary using SVM",fontdict=font)
legend_elements = [ Patch(facecolor='purple', edgecolor='k', label='Class 0'),Patch(facecolor='yellow', edgecolor='k', label='Class 1')]
plt.legend(handles=legend_elements, title="Classes")
plt.show()

print("Accuracy of SVM Model: ",model.score(x_test,y_test))
```
