```python
#Drawing a line using matplotlib
import matplotlib.pyplot as plt
import numpy as np

x_coordinates = np.array([0,3])
y_coordinates = np.array([0,3])

plt.plot(x_coordinates,y_coordinates)
plt.show()
```

```python
# Plotting a set of points on the graph
import matplotlib.pyplot as plt
import numpy as np

x_points = np.array([0,2,2,0])
y_points = np.array([0,0,2,2])
plt.plot(x_points,y_points,'o')
plt.show()
```

```python
# Drawing a polygon in the graph
import matplotlib.pyplot as plt
import numpy as np

x_coordinates = np.array([0,0,3,0])
y_coordinates = np.array([0,3,0,0])

plt.plot(x_coordinates,y_coordinates)
plt.show()
```

```python
# If we want to draw the points or a line according to a function such that the values of x get updated itself by increasing by 1(starting from 0) then we don't pass the x_points

import matplotlib.pyplot as plt
import numpy as np

y_values = np.array([0,1,4,9,16,25,36])

plt.plot(y_values)
plt.show()
```

```python
#Drawing a dashed square with red color with points as star
import matplotlib.pyplot as plt
import numpy as np

x_coor = np.array([0,2,2,0,0])
y_coor = np.array([0,0,2,2,0])

plt.plot(x_coor,y_coor,'r-.*')
plt.show()
```

```python
# Customized marker : square marker with red border blue face and size = 10

import matplotlib.pyplot as plt
import numpy as np

x_coor = np.array([0,2,2,0,0])
y_coor = np.array([0,0,2,2,0])

plt.plot(x_coor,y_coor,marker = 's',ms = 10, mfc = 'b', mec = 'r')
plt.show()

#If we want to remove the line then we use fmt of plot and don't specicy line attribute
plt.plot(x_coor,y_coor,'s',ms = 10, mfc = 'b', mec = 'r')
plt.show()
```

```python
# Drawing a customized line : Color blue dashed line and line width 10
import matplotlib.pyplot as plt
import numpy as np

y_coor = np.array([0,1,8])
plt.plot(y_coor,"r--",lw = 10)
```

```python
# Inserting labels and title in plot 

import matplotlib.pyplot as plt
import numpy as np

distance = np.array([0,2,4,9,16])
plt.plot(distance)
plt.xlabel('Time')
plt.ylabel('Distance')
plt.title("Distance-Time Graph")
plt.show()
```

```python
# Inserting customozid label and title 

import matplotlib.pyplot as plt
import numpy as np

time = np.array([0,1,2,3,4])
distance = np.array([0,2,4,9,16])


xfont = {'family': 'Times New Roman','color': 'Blue'}
yfont = {'family': 'Times New Roman','color': 'Red'}
titlefont = {'family': 'Times New Roman','color': 'Green'}
plt.plot(time,distance)
plt.xlabel("Time",fontdict = xfont)
plt.ylabel("Distance",yfont)
plt.title("Distance-Time Graph",titlefont, loc = 'center')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```
