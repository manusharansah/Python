```python
#Checking the version of matplotlib
import matplotlib
print(matplotlib.__version__)
```

```python
#Drawing a line using matplotlib
import matplotlib.pyplot as plt
import numpy as np

x_coordinates = np.array([0,3])
y_coordinates = np.array([0,3])

plt.plot(x_coordinates,y_coordinates)
plt.show()
```

```python
# Plotting a set of points on the graph
import matplotlib.pyplot as plt
import numpy as np

x_points = np.array([0,2,2,0])
y_points = np.array([0,0,2,2])
plt.plot(x_points,y_points,'o')
plt.show()
```

```python
# Drawing a polygon in the graph
import matplotlib.pyplot as plt
import numpy as np

x_coordinates = np.array([0,0,3,0])
y_coordinates = np.array([0,3,0,0])

plt.plot(x_coordinates,y_coordinates)
plt.show()
```

```python
# If we want to draw the points or a line according to a function such that the values of x get updated itself by increasing by 1(starting from 0) then we don't pass the x_points

import matplotlib.pyplot as plt
import numpy as np

y_values = np.array([0,1,4,9,16,25,36])

plt.plot(y_values)
plt.show()
```

```python
#Drawing a dashed square with red color with points as star
import matplotlib.pyplot as plt
import numpy as np

x_coor = np.array([0,2,2,0,0])
y_coor = np.array([0,0,2,2,0])

plt.plot(x_coor,y_coor,'r-.*')
plt.show()
```

```python
# Customized marker : square marker with red border blue face and size = 10

import matplotlib.pyplot as plt
import numpy as np

x_coor = np.array([0,2,2,0,0])
y_coor = np.array([0,0,2,2,0])

plt.plot(x_coor,y_coor,marker = 's',ms = 10, mfc = 'b', mec = 'r')
plt.show()

#If we want to remove the line then we use fmt of plot and don't specicy line attribute
plt.plot(x_coor,y_coor,'s',ms = 10, mfc = 'b', mec = 'r')
plt.show()
```
