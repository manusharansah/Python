```python
# Customized marker : square marker with red border blue face and size = 10

import matplotlib.pyplot as plt
import numpy as np

x_coor = np.array([0,2,2,0,0])
y_coor = np.array([0,0,2,2,0])

plt.plot(x_coor,y_coor,marker = 's',ms = 10, mfc = 'b', mec = 'r')
plt.show()

#If we want to remove the line then we use fmt of plot and don't specicy line attribute
plt.plot(x_coor,y_coor,'s',ms = 10, mfc = 'b', mec = 'r')
plt.show()
```

```python
# Inserting labels and title in plot 

import matplotlib.pyplot as plt
import numpy as np

distance = np.array([0,2,4,9,16])
plt.plot(distance)
plt.xlabel('Time')
plt.ylabel('Distance')
plt.title("Distance-Time Graph")
plt.show()
```

```python
# Inserting customozid label and title 

import matplotlib.pyplot as plt
import numpy as np

time = np.array([0,1,2,3,4])
distance = np.array([0,2,4,9,16])


xfont = {'family': 'Times New Roman','color': 'Blue'}
yfont = {'family': 'Times New Roman','color': 'Red'}
titlefont = {'family': 'Times New Roman','color': 'Green'}
plt.plot(time,distance)
plt.xlabel("Time",fontdict = xfont)
plt.ylabel("Distance",yfont)
plt.title("Distance-Time Graph",titlefont, loc = 'center')
plt.show()
```

```python
# Drawing grid lines on the plot 
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0,1,2,3,4,5,6])
y = np.array([0,1,2,3,4,5,6])

plt.plot(x,y,'o-r')
plt.grid(axis = 'both', linestyle = 'dashed', linewidth = 0.5, color = 'blue')
plt.show()
```

```python
# Subplots
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0,1,2,3])
y1 = np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x1,y1)
plt.title('SALES')

x2 = np.array([0,1,2,3])
y2 = np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.title('INCOME')

plt.suptitle('MY SHOP')

plt.show()
```

```python
# Scatter Plot in Matplotlib
import matplotlib.pyplot as plt
import numpy as np



x = np.array([0,3,4,5,2,4,6])
y = np.array([4,5,1,0,4,5,2])
sizes = np.random.randint(100,size=7)

plt.scatter(x,y,s = sizes, c= 'red', alpha = 1)
plt.show()
```

```python
# Doing more with scatter plot
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(50,size = 50)
y = np.random.randint(50,size = 50)
colors = np.random.randint(100,size=50)
size = 20 * np.random.randint(50,size = 50)

plt.scatter(x,y,c=colors,cmap='nipy_spectral',s = size, alpha = 0.7)
plt.xlabel("Lalitude")
plt.ylabel('Longitude')
plt.title('Temperature',color = 'red')
plt.colorbar()
plt.show()

```

```python
# Using bar in python using matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])

plt.bar(x,y,color='hotpink')
plt.xlabel('Car')
plt.ylabel('Cost')
plt.title('Cost - Car Graph')
plt.show()
```

```python
#Drawing a customized vertical graph of width 0.3 color red 
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])

plt.bar(x,y,color = 'red', width = 0.3)
plt.show()
```

```python
# Drawing a customized horizontal bar graph : color = black height = 0.4 
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])

plt.barh(x,y,height = 0.4, color = 'black')
plt.show()

# By default the height and the width of the horizontal and the vertical grpah is 0.8 
```

```python
#  Drawing pie-chart in python using pyplot
import matplotlib.pyplot as plt
import numpy as np

x = np.array([300,400,600,650])

plt.pie(x,startangle=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([30,45,60,45,50])
label = ['A','B','C','D','E']
plt.pie(x,autopct='%1.1f%%')
plt.legend(label,title = 'Justtocheck')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.array([200,400,600,900])
labels = ['A','B','C','D']

plt.pie(data)
plt.legend(labels, title = 'Contribution', loc = 'center right', bbox_to_anchor = (1.3,0.5))
plt.show()

# here we have used bbox-to_anchor to adjust the position of the legend 1.3 for distance from the pie chart and 0.5 for the perfect vertical alignment
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.array([200,400,600,900])
labels = ['A','B','C','D']

plt.pie(data, startangle=720)
plt.legend(labels, title = 'Contribution', loc = 'upper right', bbox_to_anchor = (1.3,1))
plt.show()
```

```python
#Also we can use the shadoe to add shadow and the explode argument to adjust the distance of the block from the center

import matplotlib.pyplot as plt
import numpy as np

data = np.array([200,400,600,900])
labels = ['A','B','C','D']
exp = [0.3,1,0,0]
plt.pie(data, shadow = True, explode = exp)
plt.legend(labels, title = 'Contribution', loc = 'center right', bbox_to_anchor = (1.4,0.5))
plt.show()
```
