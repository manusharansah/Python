# Statistics
In Machine Learning and in mathematics, while dealing with numeric data we generally come acroos three different values for knowing the characteristics of data.
1. Mean
2. Median
3. Mode


Mean is the average of the given dataset.
Median is the mid-data of teh given dataset.
Mode is the most frequent data in the given dataset.

```python
# Finding Mean
import numpy as np

dataset = [20,21,19,34,23,45,63]

mean = np.mean(dataset)
print(mean)
```

```python
# Finding the median
import numpy as np

dataset = [20,21,19,34,23,45,63]

median = np.median(dataset)
print(median)
```

```python
#Finding the modal data
#Numpy module has no method for finding median so we use the stats submodule from the scipy module to get the median method. 
 
import scipy.stats as stats
import numpy as np

dataset = np.random.randint(low = 90, high = 95 , size = 6)


mode = stats.mode(dataset)

print( mode)

# In this program mode is an Modeobject with mode value and count of the modal value


```

# Standard Deviation and Variance
Standard deviation refers to the deviation of the observations or the values form the mean value of teh given numeic dataset. More the standard deviation, higher is the non- uniformity and low spread of the data in the dataset and vice-versa. 
On the other hand, variance is the square of the standard deviation.

```python
#Finding standard deviation and variance in the given dataset
import numpy as np

dataset = np.random.normal(95,2,100)

std = np.std(dataset)
var = np.var(dataset)

print(std, var)

# In actual the std = 2 and variance = 4 
```

# Percentile in Statistics
Percentile means the value which is greater than given percenatge of data in the given dataset.
For example: if 30 percentile is 400, then it means that about 30% of the data in the given dataset are less than 400 and 70 % are greater it.

```python
# Finding percentile in a given dataset

import numpy as np

dataset = np.random.normal(170,45,500)
per30 = np.percentile(dataset, 30)

print("30 Percentile is: ", per30)
```

# Data Distribution
The spread of given datset is known as data distribution. Depending upon the nature of dataset, there are many types of data distribution such as:
1. Normal distribution (Guassian Distribution)
2. Uniform Distribution
3. Skewed Distribution

   i. Left Skewed Distribution
   
   ii. Right Skewed Distribution
and many other

The most common and useful distribution is normal distribution


```python
# Drawing a normal distribution 
import numpy as np
import matplotlib.pyplot as plt

dataset = np.random.normal(96,5,200)
plt.hist(dataset)
plt.title('Normal Distribution')
plt.show()
```

```python
# Drawing a uniform distribution 
import numpy as np
import matplotlib.pyplot as plt

dataset = np.random.uniform(55,60,200)
plt.hist(dataset)
plt.title('Normal Distribution')
plt.show()
```

# Scatter Plot
